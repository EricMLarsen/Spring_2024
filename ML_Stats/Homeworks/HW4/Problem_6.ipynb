```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import sklearn.cross_decomposition as skcd
```

```python
#Load the Dataset into program

dataset = pd.read_csv('Data/JobPerformance.csv')
print(dataset.shape)
#seperate the dataset into X and Y matrices

X_mat = dataset[['PsychTest1','PsychTest2','YrsEdu','IQ','HrsTrn','HrsWrk']]
Y_mat = dataset[['ClientSat','SuperSat','ProjCompl']]

```

```python
#Do the cross decomposition
cca = skcd.CCA(n_components=3)

cca.fit(X_mat,Y_mat)
#Get the weights
x_weights = cca.x_weights_
y_weights = cca.y_weights_

plt.figure(figsize=(10, 6))
bar_width = 0.2 

plt.bar(np.arange(len(x_weights)) - bar_width, np.abs(x_weights[:, 0]), bar_width, color='b', alpha=0.7, label='Component 1')
plt.bar(np.arange(len(x_weights)), np.abs(x_weights[:, 1]), bar_width, color='r', alpha=0.7, label='Component 2')
plt.bar(np.arange(len(x_weights)) + bar_width, np.abs(x_weights[:, 2]), bar_width, color='g', alpha=0.7, label='Component 3')

plt.xlabel('X Variables')
plt.ylabel('Weight')
plt.title('Weights of X Variables for Canonical Components')
plt.xticks(np.arange(len(x_weights)), ['PsychTest1', 'PsychTest2', 'YrsEdu', 'IQ', 'HrsTrn', 'HrsWrk'], rotation=45)
plt.legend()

plt.tight_layout()
plt.savefig('Problem6_CCA_X.png',dpi=200)
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(np.arange(len(y_weights)) - bar_width, np.abs(y_weights[:, 0]), bar_width, color='b', alpha=0.7, label='Component 1')
plt.bar(np.arange(len(y_weights)), np.abs(y_weights[:, 1]), bar_width, color='r', alpha=0.7, label='Component 2')
plt.bar(np.arange(len(y_weights)) + bar_width, np.abs(y_weights[:, 2]), bar_width, color='g', alpha=0.7, label='Component 3')

plt.xlabel('Y Variables')
plt.ylabel('Weight')
plt.title('Weights of Y Variables for Canonical Components')
plt.xticks(np.arange(len(y_weights)), ['ClientSat', 'SuperSat', 'ProjCompl'], rotation=45)
plt.legend()

plt.tight_layout()
plt.savefig('Problem6_CCA_Y.png',dpi=200)
plt.show()
```
