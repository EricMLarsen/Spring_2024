```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to be plotted
def part_i_function(x, y):
    return np.cos(x-y)
def part_ii_function(x,y):
    return np.cos(x**2 -y**2)
def part_iii_function(x,y,a,b):
    return np.tanh(a*(x*y)+b)

# Generate x and y values
x_values = np.linspace(-5, 5, 100)
y_values = np.linspace(-5, 5, 100)

# Create a grid of x and y values
x, y = np.meshgrid(x_values, y_values)

# Compute the function values for each (x, y) pair
part1_vals = part_i_function(x, y)
part2_vals = part_ii_function(x, y)
part3_vals = part_iii_function(x, y,1,1)
```

```python
#Problem 5 Part C i

# Create a heatmap using plt.pcolormesh
plt.pcolormesh(x, y, part1_vals, cmap='bwr')

# Add a colorbar for reference
plt.colorbar(label='Function Value')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heatmap of Problem 5:C subpart i')
plt.savefig('Problem5csubpart1.png',dpi=200)
```

```python
#Problem 5 Part C ii

# Create a heatmap using plt.pcolormesh
plt.pcolormesh(x, y, part2_vals, cmap='bwr')

# Add a colorbar for reference
plt.colorbar(label='Function Value')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heatmap of Problem 5:C subpart ii')
plt.savefig('Problem5csubpart2.png',dpi=200)
```
